```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
spark = SparkSession.builder.appName("DS 2.3 HW3").getOrCreate()
```

```python
df = spark.read.csv("titanic.csv", header=True, inferSchema=True)
df.show()
```

```python
df.describe().show()
```

```python
df.printSchema()
```

```python
df.where(col("Age").isNull()).count()
```

```python
def find_gender(x):
    if x == "male":
        return 1
    return 0

gender_udf = udf(find_gender, IntegerType())
df = df.withColumn("Gender", gender_udf("Sex"))
df.show()
```

```python
print("({}, {})".format(df.count(), len(df.columns)))
```

```python
for i in df.select("Age").collect():
    if i is not None:
        print(i)
```

```python
df.filter(df.Embarked == "C").show()
```

```python
df.select("Fare").describe().show()
```

```python
df.filter(df.Embarked.isNotNull()).select("Embarked").distinct().count()
```

```python
df.filter(df.Embarked.isNotNull()).groupBy("Embarked").count().show()
```

```python
df.filter(df.Sex.isNotNull()).groupBy("Sex").count().write.json("Sex_json")
```

```python
c = df.filter(df.Embarked == "C")
c_male = c.filter(c.Gender == 1)
c_female = c.filter(c.Gender == 0)
```

```python
c_male.select(mean("Age")).show()
```

```python
c_female.select(mean("Age")).show()
```

```python
c_male.select(max("Age")).show()
```

```python
c_female.select(max("Age")).show()
```
